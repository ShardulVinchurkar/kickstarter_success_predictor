```python
import numpy as np
import pandas as pd
```

```python
projects_df = pd.read_csv("data.csv", index_col=None)
projects_df.head()
```

```python
projects_df['is_success'] = projects_df['state'].apply(lambda x : 0 if x == "failed" or x == "canceled" or x == "suspended" or x == "undefined" else 1)    
```

```python
projects_df.head(n=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
label = projects_df['is_success']
df = projects_df.drop('is_success', 1)
```

```python
training_data, testing_data, training_label, testing_label = train_test_split(df, label, test_size = 0.1, stratify = label)
```

```python
training_data.head()
```

```python
training_data = training_data.drop(['ID', 'name', 'category', 'currency', 'goal', 'pledged', 'state', 'usd pledged'], axis = 1)
training_data.head()
```

```python
start_time_split = pd.DataFrame(training_data.launched.str.split(' ').tolist(), columns = ['start','time'])
frames = [training_data, start_time_split]
training_data = pd.concat([training_data.reset_index(drop=True), start_time_split.reset_index(drop=True)], axis=1)

training_data['deadline'] = pd.to_datetime(training_data['deadline'])
training_data['start'] = pd.to_datetime(training_data['start'])
set(map(type, training_data.start.values.tolist()))
set(map(type, training_data.deadline.values.tolist()))

training_data['days'] = training_data['deadline'] - training_data['start']
training_data.head()
```

```python
training_data = training_data.drop(['deadline', 'launched', 'start', 'time'], axis = 1)
training_data.head()
```

```python
training_data['days'] = training_data['days'] / np.timedelta64(1, 'D')
training_data.head()
```

```python
training_label =  training_label.drop([100221, 17183, 333239, 98007, 67616, ])
```

```python
from sklearn import linear_model
```

```python
testing_data.head()
```

```python
testing_data = testing_data.drop(['ID', 'name', 'category', 'currency', 'goal', 'pledged', 'state', 'usd pledged'], axis = 1)

start_time_split_test = pd.DataFrame(testing_data.launched.str.split(' ').tolist(), columns = ['start','time'])
frames = [testing_data, start_time_split_test]
testing_data = pd.concat([testing_data.reset_index(drop=True), start_time_split_test.reset_index(drop=True)], axis=1)

testing_data['deadline'] = pd.to_datetime(testing_data['deadline'])
testing_data['start'] = pd.to_datetime(testing_data['start'])
set(map(type, testing_data.start.values.tolist()))
set(map(type, testing_data.deadline.values.tolist()))

testing_data['days'] = testing_data['deadline'] - testing_data['start']

testing_data = testing_data.drop(['deadline', 'launched', 'start', 'time'], axis = 1)

testing_data['days'] = testing_data['days'] / np.timedelta64(1, 'D')

testing_data['normalized_days'] = stats.boxcox(testing_data['days'])[0]

testing_data['normalized_usd_goal_real'] = stats.boxcox(testing_data['usd_goal_real'])[0]

testing_data['usd_pledged_real'] += 1
testing_data['normalized_usd_pledged_real'] = stats.boxcox(testing_data['usd_pledged_real'])[0]

testing_data['backers'] += 1
testing_data['normalized_backers'] = stats.boxcox(testing_data['backers'])[0]

testing_data = testing_data.drop(['backers', 'usd_pledged_real', 'usd_goal_real', 'days'], axis = 1)

testing_data = pd.get_dummies(testing_data, columns = ['main_category', 'country'])

testing_data.head()
```

```python
from sklearn.metrics import accuracy_score
regression_model = linear_model.LinearRegression()
regression_model.fit(training_data, training_label)
test_predictions = regression_model.predict(testing_data)
```

```python
predictions = []
for n in range(len(test_predictions)):
    if test_predictions[n] < 0.4:
        predictions.append(0)
    else:
        predictions.append(1)
```

```python
accuracy = accuracy_score(testing_label, predictions)
accuracy
```
